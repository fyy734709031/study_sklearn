```python
# 练习2：餐厅小费情况分析
# 小费和总消费之间的关系（散点图+回归分析）
# 男性顾客和女性顾客，谁更慷慨（箱图或者提琴图）
# 抽烟与否是否会对小费金额产生影响（箱图或者提琴图）
# 工作日和周末，什么时候顾客给的小费更慷慨（箱图或者提琴图）
# 午饭和晚饭，哪一顿顾客更愿意给小费（箱图或者提琴图）
# 就餐人数是否会对慷慨度产生影响（箱图或者提琴图）
# 性别+抽烟的组合因素对慷慨度的影响（统计柱状图）
```

```python
import seaborn as sns
data = sns.load_dataset("tips")
data
```

```python
# 小费和总消费之间的关系（散点图+回归分析）
sns.jointplot(x='total_bill',y='tip',data=data,kind='reg')
```

```python
# 男性顾客和女性顾客，谁更慷慨（箱图或者提琴图）
sns.catplot(x='sex',y='tip', data=data, kind='box')
sns.catplot(x='sex',y='tip', data=data, kind='violin')
```

```python
# 抽烟与否是否会对小费金额产生影响（箱图或者提琴图）

sns.catplot(x='smoker',y='tip', data=data, hue = "sex", kind='box')
sns.catplot(x='smoker',y='tip', data=data, hue = "sex", kind='violin')
```

```python
# 工作日和周末，什么时候顾客给的小费更慷慨（箱图或者提琴图）

sns.catplot(x='day',y='tip', data=data, hue = "sex", kind='box')
sns.catplot(x='day',y='tip', data=data, hue = "sex", kind='violin')
```

```python
# 午饭和晚饭，哪一顿顾客更愿意给小费（箱图或者提琴图）

sns.catplot(x='time',y='tip', data=data, hue = "sex", kind='box')
sns.catplot(x='time',y='tip', data=data, hue = "sex", kind='violin')
```

```python
# 就餐人数是否会对慷慨度产生影响（箱图或者提琴图）

sns.catplot(x='size',y='tip', data=data, hue = "sex", kind='box')
sns.catplot(x='size',y='tip', data=data, hue = "sex", kind='violin')
```
