```python
import seaborn as sns
import matplotlib.pyplot as plt
# tk 在图形化gui界面中绘制图形
# inline 在浏览器中绘图
# %matplotlib tk
data = sns.load_dataset("iris")
data
```

```python
# 鸢尾花萼片（sepal）和花瓣（petal）的大小关系（散点图）
data["sepal_size"] = data["sepal_length"] * data["sepal_width"]
data["petal_size"] = data["petal_length"] * data["petal_width"]
plt.figure("鸢尾花萼片（sepal）和花瓣（petal）的大小关系",figsize=(8,6))
# ax1 = plt.subplot(221)
# 分类散点
# sns.catplot(x = "sepal_size", y = "petal_size", kind="strip", data = data)
# ax2 = plt.subplot(222)
# 分簇散点
sns.catplot(x = "sepal_size", y = "petal_size", kind="swarm", data = data)
```

```python
# 不同种类（species）鸢尾花萼片和花瓣大小的分布情况（箱图或者提琴图）
plt.figure("不同种类鸢尾花萼片和花瓣大小的分布情况",figsize=(8,6))
sns.catplot(x = "sepal_size", y = "petal_size", kind="box", hue = "species", data = data)
sns.catplot(x = "sepal_size", y = "petal_size", kind="violin", hue = "species", data = data)
sns.catplot(x = "sepal_size", y = "petal_size", kind="boxen", hue = "species", data = data)

#     plt.boxplot(sepal_size.values
#                 ,patch_artist = True
#                # 中位数线颜色
#                , medianprops = {'color': 'b'}
#                # 箱子颜色设置，color：边框颜色，facecolor：填充颜色
#                , boxprops = {'color': 'b', 'facecolor': 'r'}
#                # 猫须颜色whisker
#                , whiskerprops = {'color': 'r'}
#                # 猫须界限颜色whisker cap
#                , capprops = {'color': 'b'})
# 配对网格
# g = sns.PairGrid(data,
#                  x_vars=["species"],
#                  y_vars=["sepal_size", "petal_size"],
#                  aspect=2, size=4)
# g.map(sns.violinplot, palette="pastel");
```

```python
# seaborn.jointplot(x, y, data=None, kind='scatter', stat_func=None, 
    # color=None, height=6, ratio=5, space=0.2, dropna=True, xlim=None, 
    # ylim=None, joint_kws=None, marginal_kws=None, annot_kws=None, **kwargs)
#     kind =  { “scatter”-散点图 | “reg”-线性回归 | “resid” | “kde”-六角箱图 | “hex”-核密度估计图 }
# http://seaborn.pydata.org/generated/seaborn.jointplot.html#seaborn.jointplot
    
# 鸢尾花萼片和花瓣大小的联合分布情况（六角箱图或者核密度估计）
# 六角箱图
sns.jointplot(x='sepal_size',y='petal_size',data=data,kind='kde')
# 核密度估计图
sns.jointplot(x='sepal_size',y='petal_size',data=data,kind='hex')
```
