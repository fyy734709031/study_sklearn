```python
import pandas as pd
```

```python
# 1、加载数据（users.csv）
users_csv = pd.read_csv("./users.csv", sep = "|")
users_csv
```

```python
# 2、以occupation分组，求每一种职业所有用户的平均年龄；
users_csv.groupby("occupation")["age"].mean()
```

```python
# 3、求每一种职业男性的占比，作为新的一列(male_pct)添加到数据集中,并按照从低到高的顺序排列；
male_pct = users_csv.groupby("occupation")["gender"].value_counts(normalize=True)
male_pct.unstack()["M"]
# 作为新的一列(male_pct)添加到数据集中
```

```python
# 4、获取每一种职业对应的最大和最小的用户年龄
maxAge = users_csv.groupby("occupation")["age"].max()
minAge = users_csv.groupby("occupation")["age"].min()
max_minAge = pd.merge(maxAge, minAge, left_on = "occupation", right_on = "occupation", suffixes = ["_max", "_min"])
max_minAge
```
